```python
import pandas as pd
```

```python
# Extracting file #1 for project
ctrys_df = pd.read_csv("countries.csv")
ctrys_df.head()
```

```python
# Transform Data to remove ambiguous columns
ctrys_df = ctrys_df.rename_axis('id', axis='columns')
countries = ctrys_df.drop(columns=['Service','Industry', 'Climate'])

# Make sure no spaces are interfering with our connecting data (spoiler: They are right now!)
countries['Country'] = countries['Country'].str.replace(" ","")
countries.head()
```

```python
countries.to_csv('countries_updated.csv', index=False)
```

```python
# Extract second dataset for transformation
happy = pd.read_csv('happiness.csv')
happy_df = pd.DataFrame(happy)
happy_df.head()
```

```python
# Transform second dataset to remove ambiguous columns or those not defined
# in dataset to make our database cleaner
happy_updated = happy_df.drop(columns=['Whisker.high', 'Whisker.low'])
happy_updated['Country'] = happy_updated['Country'].str.strip()

# Finish Transformation and prepare for SQL loading
happy_updated.to_csv('happiness_updated.csv', index=False)
```
